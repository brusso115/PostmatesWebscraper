```python
import pandas as pd
import numpy as np
import re
```

```python
postmates = pd.read_csv('/Users/baileyrusso/PycharmProjects/GrubHubScraper/test3.csv',header=None)
postmates.columns = ['Name', 'Category', 'Favorites', 'Address', 'MenuItems', 'PopularItems']
```

# Clean Address Column

```python
addressList = []
for i in range(len(postmates.Address)):
    address = postmates.Address.iloc[i]
    if pd.notnull(address):
        addressList.append(address.split('\n')[1].replace('(','').replace(')','').rstrip())
    else:
        addressList.append(np.nan)
        
postmates.Address = addressList
```

# Clean Favorites Column

```python
postmates['Favorites'] =list(map(lambda row: str(row).replace('added to favorites', '').rstrip(), postmates['Favorites']))

```

```python
for i in range(len(postmates['MenuItems'])):
    postmates['MenuItems'][i] = list(map(lambda item: re.split(",(?=(?:[^\']*\'[^\']*\')*[^\']*$)", item), postmates['MenuItems'][i]))
```
